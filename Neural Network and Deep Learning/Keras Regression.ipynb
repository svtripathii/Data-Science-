```python
df= df.drop('id', axis=1)
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
df['year'] = df['date'].apply(lambda date: date.year) #lambda  function to extract month from 'date' column def year_extraction(date): return date.year
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
df.head()
```

```python
sns.boxplot(x='month', y='price', data=df)
```
